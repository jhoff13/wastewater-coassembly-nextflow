```python
import pandas as pd
import ast
```

```python
#import libraries csv, format/ filter.

reads_DF = pd.read_csv('/Users/jhoff/Desktop/esvelt_lab/sheets/Rothman_libraries.csv')

reads_DF = reads_DF[reads_DF['sample'].str.contains('HTP')].reset_index(drop=True)
reads_DF = reads_DF.merge(reads_DF['sample'].str.split('_',expand=True).drop([0,4],axis=1)\
        .rename({1:'month',2:'day',3:'year',5:'processing'},axis=1),
              left_index=True, right_index=True)
reads_DF['processing'] = reads_DF.processing.fillna('enriched')
reads_DF['year'] = reads_DF.year.str.replace(r'^21','2021',regex=True)
reads_DF['datetime'] = pd.to_datetime(reads_DF[['year','month','day']])
reads_DF = reads_DF.sort_values(['datetime'])
reads_DF
```

```python
#filter for unenriched and add paths:

unenriched_reads_DF = reads_DF[reads_DF['processing']=='unenriched'].reset_index(drop=True)
unenriched_reads_DF['S3_paths'] = unenriched_reads_DF['library']\
    .apply(lambda x: f'{x}_bbduk_noribo_XXX.fastq.gz')\
    .str.replace('XXX','{1/2}')
#included s3://nao-mgs-workflow-scratch/Rothman-PRJNA729801/output-unenriched/preprocess/ribo_initial/ in config
unenriched_reads_DF
```

```python
#format for coassembly

coassembly_iteration = 10
coassembly_list = []

i=coassembly_iteration
while i > 0:
    coassembly_list.append(list(range(0,i)))
    i-=1
    
out_library_DF = pd.DataFrame({'library':list(unenriched_reads_DF['S3_paths'][:coassembly_iteration])\
                               , 'coassembly': coassembly_list})

read_path = 's3://nao-mgs-workflow-scratch/Rothman-PRJNA729801/output-unenriched/preprocess/ribo_initial/'
out_library_DF['library'] = read_path + out_library_DF.library

out_library_DF
```

```python
out_library_DF.to_csv('/Users/jhoff/Documents/assembly_nextflow/libraries.csv',index=None)
```
